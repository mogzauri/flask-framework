### Example Plot of Stock Price

Example plot using matlib, seaborn, and bokeh. First reading in the csv of wiki prices created. Start by importing libraries. 

```python
# import pandas
import pandas as pd
import numpy as np

# bokeh plots
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
```

```python
# read csv
wiki_stock = pd.read_csv('./data/wiki_stock_price.csv')
```

```python
# print head
wiki_stock.head()
```

Subsetting just ticker A to plot timeseries data.

```python
# subset to specific ticker
ticker_df = wiki_stock.loc[wiki_stock['ticker'] == 'A']
# view first rows
ticker_df.head()
```

Define datetime range.

```python
def datetime(x):
    return np.array(x, dtype=np.datetime64)
```

```python
# check to see if datetime works
datetime(ticker_df['date'])
```

Test figure of Ticker A open and close. I will plot bokeh figure inline, but I'm not sure it can save since it looks like it keeps running.

```python
# plots figure inline
output_notebook()
```

Create simple line plot.

```python
# create figure
p = figure(x_axis_type = 'datetime', title = 'Stock Closing Price')
# define some options
p.grid.grid_line_alpha=0.3
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Prince'


# add line of ticker 
p.line(datetime(ticker_df['date']), ticker_df['adj_close'], color = 'blue', legend = 'Close')
p.line(datetime(ticker_df['date']), ticker_df['adj_open'], color = 'black', legend = 'Open')
p.legend.location = 'top_left'
```

Show plot.

```python
show(p)
```
